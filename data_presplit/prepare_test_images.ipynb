# Prepare test images

This module will take in a test folder, which contains subfolders for each class.

We will flatten the directory structure while prepending the folder name to the new filename.

This can then be used in the test routines to extract out the ground truth for ease.

```python
import os
from pathlib import Path
from shutil import copy
```

```python
SRC_FOLDER = os.path.join('data_224_split', 'test')
DST_FOLDER = os.path.join('data_224_test_combined')

Path(DST_FOLDER).mkdir(parents=True, exist_ok=True)
```

```python
SRC_FOLDER, DST_FOLDER
```

```python
counter = 1

for root, dirs, files in os.walk(SRC_FOLDER):
    for file in files:
        dir_label = os.path.basename(root)
        new_filename = dir_label + '___' + '{0:04}'.format(counter) + os.path.splitext(file)[-1]
        counter += 1
        
        # copy from source to destination
        src_file = os.path.join(root, file)
        dst_file = os.path.join(DST_FOLDER, new_filename)
        print('.', end='')
        copy(src_file, dst_file)
```
